- apply 함수
 - 어떠한 함수를 Series의 각각의 element에 적용시켜주는 함수
 - map 함수 보다 적용할 수 있는 함수의 범위가 넓음.
 - series.apply(func,args,**kwds)

```python
s = pd.Series([20,21,12],
              index =['london', 'New York', 'Helsinki'])
```

* args(argument의 줄임말)
** kwargs  (keyword argument의 줄임말)

```python
* args(argument의 줄임말)
```

```python
def sub_custom_value(x,val):
  return x-val
  
s.apply(sub_custom_value, args = (10,))  #arg : 특정열에서 뭘 바꾸고 싶을 때 
```

```python
def add_custom_values(x, **kwargs):
  for month in kwargs:
    x+=kwargs[month]   #month가 매개변수에 없는데 넣음, 기준변수가 없음음
  return x

# 이 함수가 좋지 않은 이유?
```

```python
s.apply(add_custom_values, june=30, july=20, august=25)

#결과값이 훨씬 더 크게 나옴 
#args가 같던 말던 그냥 합이 다 더해짐 
```

```python
def add_custom_values(x, **kwargs):
  for month in kwargs:
    x+=kwargs 
  return x
```

- merge
 - key를 이용해 데이터 row를 기준으로 연결시켜 합침(sql의 join과 유사)
 - how : 어떤 방식으로 병합할 것인가('inner', 'outer','left','right')
 - on : 어떤 label을 기준으로 병합할 것인가, 기본적으로 중복되는 칼럼을 기준으로 병합함.

```python
pd.merge(data1, data2, on='id')
```

```python
pd.merge(data1, data2, on='id', how='inner')
```

```python
pd.merge(data1, data2, on='id', how='outer')
```

```python
pd.merge(data1, data2, on='id', how='left')
```
