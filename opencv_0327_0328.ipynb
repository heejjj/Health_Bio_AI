```python
def genGausKernel1D(width, sigma):
    arr = np.arange((width//2)*(-1), (width//2)+1) 
    
    #중심에서부터의 거리를 계산함,
    arr = np.array([float(x) for x in arr])
    
    #float 처리를 해주지 않으면 값이 달라질 수 있어서 처리
    
    kernel_1d= np.exp((-arr*arr)/(2*sigma**2)) 
    
    #위에서 봤던 수식을 그대로 적용했다. 달라진 점은, 앞에 상수로 나눠주는 부분을 생략한건데 
    #어차피 나중에 커널 전체 값의 합을 1로 만들기 위해서 커널 전체의 합으로 나눠주기 때문에 
    #상수부분이 필요 없다고 생각해서 나누지 않았다
    
    kernel_1d /= kernel_1d.sum()
    #kernel은 모든 element의 합이 1이어야하기 때문에 전체 수의 합으로 나눠준다. 
    #비율이 중요하다는 뜻으로 해석했다.
    
    
    return np.array([kernel_1d])
```

```python
#Gaussian noise
def gaussian_noise(src, std):

  height, width = src.shape
  result = np.zeros((height, width), dtype='uint8')

  for i in range(height):
    for j in range(width):
      #정규분포 (가우시안 곡선)
        noise = std*np.random.normal()   #stdrk 5이면 노이즈가 더 심하게 낌
        #random값은 -값까지 다 있음, -값이 들어가 있으면 overflow가 생길 수도 있음음
        temp = src[i,j] + noise
        temp = np.clip(temp,0, 255) #overflow를 해결하기 위해 clip(자름름)
        result[i,j] = temp    
 
  return result
```

```python
import math

def PSNR(img1, img2):
  img1 = img1.astype(float)/255,
  img2 = img2.astype(float)/255,
  mes = np.mean((img1 - img2) **2)
  if mes == 0:
    return 10000000000
  return (10* math.log(1. /mes))
```

```python
#라플라시안 필터:화질 개선 시 쓰임

def laplacian_filter(src):

  src = src.astype(float)
  kernel = np.array([[0,-1,0], [-1,4,-1], [0, -1, 0]], dtype=float)
  res = cv2.filter2D(src, -1, kernel); #-1 : flag 영상외곽까지 알아서 처리리
  res = res.astype(float)

  return res
```
