```python
def CLAHE(img, th):
    height, width = img.shape
    res = np.zeros((height, width), dtype='uint8')
    cnt = np.zeros(256, dtype='uint8')   # 256 dimension  array

    histogram = np.zeros(256, dtype='uint8')

    # step1. Histogram
    for i in range(height):
        for j in range(width):
            histogram[img[i,j]] += 1

    # step2. CDF
    pdf = histogram/(width * height)

    area = 0

    for i in range(256):
        if (pdf[i] > th):
            area += (pdf[i] - th)
            pdf[i] = th
        else:
            pdf[i] = pdf[i]

    pdf += area/256

    cdf = np.zeros(256, dtype=float)

    cdf[0] = pdf[0]
    for i in range(1, 256):
        cdf[i] = pdf[i] + cdf[i-1]

    trans_f = np.round(cdf * 255)

    for i in range(height):
        for j in range(width):
            res[i,j] = trans_f[img[i,j]]
            
    return pdf, cdf, res
```

```python
# Image resizing ( )
# nearest neighbor
def nearest_neighbor(img, scale_x, scale_y):
    height, width = img.shape

    re_height = (int)(scale_y * height);
    re_width = (int)(scale_x * width);
    res = np.zeros((re_height, re_width), dtype='uint8')
    for i in range(re_height):
        for j in range(re_width):
            pi = (int)(np.floor(i / scale_y))
            pj = (int)(np.floor(j / scale_x))
            res[i, j] = img[pi, pj]
    
    return res
```

```python
# Image resizing ( )
# bilinear interpolation
def bilinear_interpolation(img, scale_x, scale_y):

    height, width = img.shape
    img = img.astype(float)
    re_height = (int)(scale_y * height);
    re_width = (int)(scale_x * width);
    res = np.zeros((re_height, re_width), dtype=float)
    
    print(res.shape)
    for i in range(re_height - scale_y):
        for j in range(re_width - scale_x):
            pi = i / scale_y
            pj = j / scale_x

            # left top
            pi_lt = (int)(np.floor(pi))
            pj_lt = (int)(np.floor(pj))
            h_lt = 1 - (pi - pi_lt)
            w_lt = 1 - (pj - pj_lt)
            area_lt = h_lt * w_lt

            # right top
            h_rt = 1 - (pi - pi_lt)
            w_rt = 1 - w_lt
            area_rt = h_rt * w_rt

            # left bottom
            h_lb = 1 - h_lt
            w_lb = w_lt
            area_lb = h_lb * w_lb

            # right bottom
            h_rb = 1 - h_lt
            w_rb = 1 - w_lt
            area_rb = h_rb * w_rb

            res[i, j] += area_lt*img[pi_lt, pj_lt]
            res[i, j] += area_rt*img[pi_lt, pj_lt+1]
            res[i, j] += area_lb*img[pi_lt+1, pj_lt]
            res[i, j] += area_rb*img[pi_lt+1, pj_lt+1]
    res = res.astype('uint8')
    return res
```

```python
#zero padding 

def image_padding(src, radius):
    height, width = src.shape
    res = np.zeros((height+2*radius, width+2*radius), dtype='uint8')
    
    for i in range(width+2*radius):
        for j in range(height+2*radius):
            pi = i - radius
            pj = j - radius
            
            if(pi >= 0 and pi < height and pj >= 0 and pj < width):
                    res[i, j] = src[pi, pj]
    return res
```

```python
def average_filter(src, radius):
    
    src = src.astype(int)
    height, width = src.shape
    avg_blur = np.zeros((height, width), dtype='uint8')
    
    for i in range(radius, height-radius):
        for j in range(radius, width-radius):
            patch = src[i-radius:i+radius+1, j-radius:j+radius+1]
            avg_blur[i, j] = np.mean(patch)
            
    return avg_blur
```

```python
def mirror_padding(src, radius):
    
    height, width = src.shape
    res = np.zeros((height+2*radius, width+2*radius), dtype='uint8')
    
    for i in range(width+2*radius):
        for j in range(height+2*radius):
            pi = i - radius
            pj = j - radius
            
            if(pi < 0):
                pi = 0
            elif(pi > height-1):
                pi = height-1
            
            if(pj < 0):
                pj = 0
            elif(pj > width-1):
                pj = width-1
            
            res[i, j] = src[pi, pj]
    return res
```
